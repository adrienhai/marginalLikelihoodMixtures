#### Compute the estimators of the marginal likelihood

```python
df1_list=[]
repetitions=20
Size=['Small','Medium','Full']
for size in Size:
    j=0
    print('\n',size,'-size','galaxies data set : ')
    hyperParams='hyperParameters'+str(size)
    methods='methods'+str(size)
    data='gal'+str(size)
    for method in eval(methods):
        for l in range(repetitions):
            i=(j+1)/(repetitions*len(eval(hyperParams)))*100
            sys.stdout.write('\r')
            sys.stdout.write("[%-25s] %d%%" % ('='*int(i/4), i))
            sys.stdout.flush()
            sleep(0.25)
            func = getattr(eval(data), eval(methods)[method])
            marginal=func(**eval(hyperParams)[method])
            df1_list.append([marginal,method,size])
            j+=1

    
```

#### Plot

```python
df1=pd.DataFrame(df1_list, columns=['marginal', 'method', 'Size'])
```

```python

sns.set(font_scale = 1.6,style='white')

g=sns.catplot(x="method", y="marginal", data=df1, col="Size", kind="box", sharex=False,sharey=False, hue="method", dodge=False,
              width=0.6,legend=True,col_wrap=1,height=8.27, aspect=10/8.27)
g.set_xticklabels(rotation=60)
g.fig.tight_layout()
g.set_ylabels('$\widehat {m(y)}$')
g.set_xlabels('')

g.axes[0].axhline(np.mean(df1[(df1['method']=='arithmeticMean') & (df1['Size']=='Small')]['marginal']), ls=':',color='black',alpha=0.3)
g.axes[1].axhline(np.mean(df1[(df1['method']=='arithmeticMean') & (df1['Size']=='Medium')]['marginal']), ls=':',color='black',alpha=0.3)
g.axes[2].axhline(np.mean(df1[(df1['method']=='arithmeticMean') & (df1['Size']=='Full')]['marginal']), ls=':',color='black',alpha=0.3)


g.axes[0].set_title('$Size=small$')
g.axes[1].set_title('$Size=medium$')
g.axes[2].set_title('$Size=full$')

```
